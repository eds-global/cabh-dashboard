```python
import matplotlib
from matplotlib.cm import ScalarMappable
from matplotlib.colors import Normalize

matplotlib.rcParams['font.family'] = 'Arial'
data = pd.read_csv(r"C:\Users\abhis\Calender\System.csv", parse_dates=['datetime'], dayfirst=True)
data['datetime'] = pd.to_datetime(data['datetime'], format='%d-%m-%Y %H:%M', errors='coerce')

data.set_index('datetime', inplace=True)
start_date = pd.to_datetime("2024-04-19")
end_date = pd.to_datetime("2024-11-10")
filtered_data = data[(data.index >= start_date) & (data.index <= end_date)]

features = ['aqi', 'pm25', 'pm10', 'co2', 'voc', 'temp', 'humidity', 'battery', 'viral_index']

if filtered_data.empty:
    print(f"No data available for the selected date range. Please check the date formatting.")
else:
    fig, axes = plt.subplots(nrows=len(features), ncols=1, figsize=(16, 8 * len(features)))

    if len(features) == 1:
        axes = [axes]

    

    for i, feature in enumerate(features):
        daily_data = filtered_data[feature].resample('D').mean()

        vmin = daily_data.min()
        vmax = daily_data.max()

        im = calplot.yearplot(daily_data,
                              cmap='Wistia',
                              edgecolor = 'black',
                              linewidth=1,              
                              ax=axes[i])               
        axes[i].set_title(f'{feature} Calendar Heatmap', fontsize=16)

        norm = Normalize(vmin=vmin, vmax=vmax)
        sm = ScalarMappable(norm=norm, cmap='Wistia')

        cbar = fig.colorbar(sm, ax=axes[i], orientation='vertical', fraction=0.02, pad=0.04)
        cbar.set_label(f'{feature} Level')  

    plt.tight_layout()

    plt.show()
```
